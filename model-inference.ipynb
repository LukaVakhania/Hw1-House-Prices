```python
test = pd.read_csv("/kaggle/input/house-prices-advanced-regression-techniques/test.csv")

cols_to_drop = ['Alley', 'PoolQC', 'Fence', 'MiscFeature', 'Id']
test = test.drop(columns=cols_to_drop, errors='ignore')

num_cols = test.select_dtypes(include='number').columns
cat_cols = test.select_dtypes(include='object').columns

test[num_cols] = test[num_cols].fillna(test[num_cols].mean())
test[cat_cols] = test[cat_cols].fillna('None')
test = pd.get_dummies(test)
```

```python
import mlflow
import mlflow.sklearn

mlflow.set_tracking_uri("https://dagshub.com/LukaVakhania/Hw1-House-Prices.mlflow")

run_id = "cafb3d0b15874f83bdffc34b6505d27f"
model_uri = f"runs:/{run_id}/model"
model = mlflow.sklearn.load_model(model_uri)
```

```python
predictions = model.predict(test)

submission = pd.DataFrame({
    "Id": test["Id"],
    "SalePrice": predictions
})
submission.to_csv("submission.csv", index=False)
```
